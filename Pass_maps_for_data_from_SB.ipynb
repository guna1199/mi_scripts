```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Arc
import json
```

Importing Dataset of BEL_JPN match


```
with open('69301.json') as f:
    df = json.load(f)
engswd = pd.json_normalize(df,sep='_')
engswd.head()
```

Passes dataframe Creation

```
passes = engswd.loc[engswd['type_name'] == 'Pass']
passes.head()
```

```
pitch_creation()
ax = plt.gca()
for i,passe in passes.iterrows():
        x = passe['location'][0]
        y = passe['location'][1]

        if passe['team_name'] == "England Women's":
            COLOR = 'red'
            pass_pt = plt.Circle((x,80-y),radius=3,facecolor=COLOR)
            pass_pt.set_alpha(0.2)
        else:
            COLOR = 'blue'
            pass_pt = plt.Circle((120-x,y),radius=3,facecolor=COLOR)
            pass_pt.set_alpha(0.2)

        ax.add_patch(pass_pt)
```

#PITCH CREATION

```
def pitch_creation():

    #OUTLINE OF PITCH
    fig = plt.figure()
    axes = fig.add_subplot(111)
    plt.plot([0,0],[0,90],color = 'white')
    plt.plot([0,130],[90,90],color = 'white')
    plt.plot([130,130],[0,90],color = 'white')
    plt.plot([0,130],[0,0],color = 'white')
    plt.plot([65,65],[0,90],color = 'black')

# INNER CIRCLES PLOT

    axes = plt.subplot()
    center_circle = plt.Circle((65,45), radius=9,color = 'white',fill = False)
    center_mark = plt.Circle((65,45), radius = 1, color = 'black')
    left_penalty_spot = plt.Circle((14,44),radius = 0.8, color = 'black')
    right_penalty_spot = plt.Circle((116,44),radius = 0.8, color = 'black')
    axes.add_patch(center_circle)
    axes.add_patch(center_mark)
    axes.add_patch(left_penalty_spot)
    axes.add_patch(right_penalty_spot)

# Left Penalty Area
    plt.plot([0,0],[20,68],color = 'white')
    plt.plot([0,20],[20,20],color = 'white')
    plt.plot([0,20],[68,68],color = 'white')
    plt.plot([20,20],[20,68],color = 'white')

# Left 6-yard BOX
    plt.plot([0,0],[30,57],color = 'white')
    plt.plot([10,0],[30,30],color = 'white')
    plt.plot([10,10],[30,57],color = 'white')
    plt.plot([10,0],[57,57],color = 'white')

# Right Penalty Area
    plt.plot([130,130],[20,68],color = 'white')
    plt.plot([110,130],[20,20],color = 'white')
    plt.plot([130,110],[68,68],color = 'white')
    plt.plot([110,110],[20,68],color = 'white')

# Right 6-Yard Box
    plt.plot([130,130],[30,57],color = 'white')
    plt.plot([120,130],[30,30],color = 'white')
    plt.plot([120,120],[30,57],color = 'white')
    plt.plot([120,130],[57,57],color = 'white')

# ARCS
    left_arc = Arc((15,45),width=18.6,height=18.6,angle=0,theta1=300.8,theta2=60,color= 'white')
    right_arc = Arc((115,45),width=18.6,height=18.6,angle=0,theta1=125,theta2=240,color= 'white')
    leftd_corner = Arc((0,0),width= 6, height= 6,angle= 0, theta1= 360,theta2=90,color= 'white')
    leftup_corner = Arc((0,90),width= 6, height= 6,angle= 0, theta1= 260,theta2=360,color= 'white')
    rightup_corner = Arc((130,90),width= 6, height= 6,angle= 0, theta1= 180,theta2=270,color= 'white')
    rightd_corner = Arc((130,0),width= 6, height= 6,angle= 0, theta1= 90,theta2=180,color= 'white')

    axes.add_patch(leftd_corner)
    axes.add_patch(leftup_corner)
    axes.add_patch(rightd_corner)
    axes.add_patch(rightup_corner)

    axes.add_patch(left_arc)
    axes.add_patch(right_arc)

#Left GOAL POST
    plt.plot([0,0],[37,52],color = 'white')
    plt.plot([-7,-7],[37,52],color = 'white')
    plt.plot([-7,0],[37,37],color = 'white')
    plt.plot([-7,0],[52,52],color = 'white')

#RIGHT GOAL POST
    plt.plot([130,130],[37,52],color = 'white')
    plt.plot([137,137],[37,52],color = 'white')
    plt.plot([137,130],[37,37],color = 'white')
    plt.plot([137,130],[52,52],color = 'white')


    plt.axis('off')
```

# PASSES BY CAROLINE SEGER

```
carol = engswd.loc[engswd['player_name'] == 'Sara Caroline Seger']
pitch_creation()
ax = plt.gca()
for i,c_pass in carol.iterrows():
    if c_pass['type_name'] == 'Pass':
        x = c_pass['location'][0]
        y = c_pass['location'][1]

        dx = c_pass['pass_end_location'][0]
        dy = c_pass['pass_end_location'][1]

        pass_circle = plt.Circle((x,90-y),radius=3,color='blue')
        pass_circle.set_alpha(0.4)
        ax.add_artist(pass_circle)

        pass_arrow = plt.Arrow(x,90-y,dx-x,dy-y,width = 4,color='blue')
        ax.add_artist(pass_arrow)
```
